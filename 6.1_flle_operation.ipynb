#### File Operation- Read And write Files
File handling is a crucial part of any programming language. python provide built-in functions and method to
read from and write, both text and binary. This lesson will cover the basics of file handling, include reading
and writing text and binary files.

```python
## Read a whole file
with open("example.txt",'r') as file:
    for line in file:
        print(line)
```

```python
with open("example.txt",'r') as file:
    for line in file:
        print(line.strip()) ## strip() remove the newline charevter
```

```python
## writing a file(Overriting)
with open("example.txt",'w')as file:
    file.write("Hii this is riyanshu from varanasi\n")
    file.write("how are you ")
    
 
```

```python
## Write a file without overriding
with open("example.txt",'a') as file:
    file.write("\nAppend operation taking place\n")
```

```python
## Wrinting a list of line to a file
lines=["First line\n","second line\n","thard line"]
with open("example.txt",'a') as file:
    file.writelines(lines)
```

```python
## ,Binary files
# Writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open("example.bin",'wb') as file:
    file.write(data)
```

```python
### Read a text file and count thee number of lines, words, and characters.
# Conting line,words, and charecters in a text file.
def count_text(file_path):
    with open("example.txt",'r') as file:
        lines=file.read()
        count_line=(len(lines))
        count_words=sum((len(lines) for line in lines))
        count_char=sum((len(lines)for line in lines))
    return count_line,count_char,count_words
file_path="example.txt"
lines,words,chareter=count_text(file_path)
print(f"lines:{lines},\nwords:{words},\ncharecter{chareter}")
```
